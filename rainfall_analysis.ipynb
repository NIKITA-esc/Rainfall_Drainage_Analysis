```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall_df = pd.read_excel('Chd_Rainfall_Drainage.xlsx', sheet_name = 'monthly_rainfall')
drainage_df = pd.read_excel('Chd_Rainfall_Drainage.xlsx', sheet_name = 'drainage_capacity')
```

```python
df_2025 = drainage_df[(drainage_df['Year'] == 2025) & (~drainage_df["Month_Name"].isin(["Oct", "Nov", "Dec"]))]
```

```python
rainfall_df
```

```python
total_rainfall = rainfall_df.groupby('Year')['Total_Rainfall_mm'].sum()
```

```python
total_rainfall
```

```python
total_rainfall_area_2025 = df_2025.groupby('area_name')['Estimated_area_rain_mm'].sum()
```

```python
total_rainfall_area_2025
```

```python
# 1ï¸âƒ£ Total rainfall per year
plt.figure(figsize=(10, 5))
total_rainfall.plot(marker='o', color='blue')
plt.title("Total Rainfall per Year (2018â€“2025)")
plt.xlabel("Year")
plt.ylabel("Total Rainfall (mm)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# 2ï¸âƒ£ Total estimated rainfall by area (2025)
plt.figure(figsize=(10, 6))
total_rainfall_area_2025.plot(kind="bar", color="teal")
plt.title("Estimated Rainfall by Area (2025)")
plt.xlabel("Area")
plt.ylabel("Estimated Rainfall (mm)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python

# Filter rainfall data up to August
drainage_upto_aug = df_2025[~df_2025["Month_Name"].isin(["Sep", "Oct", "Nov", "Dec"])]

# 2ï¸âƒ£ Total estimated rainfall by area (2025, up to Aug only)
total_rainfall_area_2025_aug = drainage_upto_aug.groupby("area_name")["Estimated_area_rain_mm"].sum()

plt.figure(figsize=(10, 6))
total_rainfall_area_2025_aug.plot(kind="bar", color="teal")
plt.title("Estimated Rainfall by Area (2025, Up to Aug)")
plt.xlabel("Area")
plt.ylabel("Estimated Rainfall (mm)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
